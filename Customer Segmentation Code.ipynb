```python
x1=df.loc[:,["Distance(km)","PurchaseRate"]].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
  kmeans=KMeans(n_clusters=k,init="k-means++")
  kmeans.fit(x1)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.grid()
plt.plot(range(1,11),wcss,linewidth=2,color="red",marker="8")
plt.xlabel("K Value")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5)

label=kmeans.fit_predict(x1)

print(label)

```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color="black")
plt.title("Clusters of Customers")
plt.xlabel("Distance(km)")
plt.ylabel("PurchaseRate")
plt.show()
```

```python
clusters= kmeans.fit_predict(x1)
df["label"]=clusters

from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure(figsize=(20,10))
ax=fig.add_subplot(111, projection='3d')
ax.scatter(df.Age[df.label==0], df["Distance(km)"][df.label==0], df["PurchaseRate"][df.label==0], c='blue',s=60)
ax.scatter(df.Age[df.label==1], df["Distance(km)"][df.label==1], df["PurchaseRate"][df.label==1], c='red',s=60)
ax.scatter(df.Age[df.label==2], df["Distance(km)"][df.label==2], df["PurchaseRate"][df.label==2], c='green',s=60)
ax.scatter(df.Age[df.label==3], df["Distance(km)"][df.label==3], df["PurchaseRate"][df.label==3], c='orange',s=60)
ax.scatter(df.Age[df.label==4], df["Distance(km)"][df.label==4], df["PurchaseRate"][df.label==4], c='purple',s=60)
ax.view_init(30,185)
plt.xlabel("Age")
plt.ylabel("Distance(km)")
ax.set_zlabel("PurchaseRate")

plt.show()
```
